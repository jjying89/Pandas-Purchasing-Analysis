```python
#import dependencies
import pandas as pd

#find file
file_to_load = "Resources/purchase_data.csv"

#read file and store into pd
purchase_data = pd.read_csv(file_to_load)
```

```python
#rename df to hop(heroes of pymoli)
hop = purchase_data
hop
```

## Player Count

* Display the total number of players

```python
#count total unique players
totalplayers = hop['SN'].nunique()

#print result
totalplayers_df = pd.DataFrame({"Total Players":[totalplayers]})
totalplayers_df
```

## Purchasing Analysis (Total)
* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
#purchasing anaylysis (total)
numberofuniqueitems = hop['Item Name'].nunique()
averagepurchaseprice = hop['Price'].mean()
totalnumberofpurchases = hop['Purchase ID'].count()
totalrevenue = hop['Price'].sum()

#print results
purchasinganalysis_total_df = pd.DataFrame({"Number of Unique Items":[numberofuniqueitems],
                                        "Average Price":[averagepurchaseprice],
                                        "Number of Purchases":[totalnumberofpurchases],
                                        "Total Revenue":[totalrevenue]})
purchasinganalysis_total_df
```

## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed

```python
#gender demographics
#drop duplicate SN to get unique total players
gender_df = hop.drop_duplicates(subset='SN', keep="first")

#find & print # of players for each gender
gender_df.groupby(['Gender']).SN.count()
```

```python
#gender demographics
#assign number of players by gender
maleplayers = 484
femaleplayers = 81
otherplayers = 11

#find percentage of each gender
percentage_male = maleplayers / totalplayers * 100
percentage_female = femaleplayers / totalplayers * 100
percentage_other = otherplayers / totalplayers * 100

#rename rows and print results
gender_demo_df = pd.DataFrame({"Total Count": [484, 81, 11],
    "Percentage of Players": [percentage_male, percentage_female, percentage_other]})
gender_demo_df.rename(index={0:'Male',1:'Female',2:'Other/Non-Disclosed'}, inplace=True)
gender_demo_df
```

## Purchasing Analysis (Gender)

* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender




* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
#puchasing analysis (gender)
#remove spaces in column name
hop.columns = [c.replace(' ', '_') for c in hop.columns]

#find purchase count, avg purchase price, total purchase value, and avg total purchase per person by gender
purchasecount = hop.groupby(['Gender']).Purchase_ID.count()
avgpurchaseprice = hop.groupby(['Gender']).Price.mean()
totalpurchasevalue = hop.groupby(['Gender']).Price.sum()
avgtotalpurchaseperperson = totalpurchasevalue / [femaleplayers, maleplayers, otherplayers]
```

```python
#puchasing analysis (gender)
#merge groupby into df
purchasinganalysis_gender_df = pd.merge(purchasecount, avgpurchaseprice, on="Gender", how="outer")
purchasinganalysis_gender_df = pd.merge(purchasinganalysis_gender_df, totalpurchasevalue, on="Gender", how="outer")
purchasinganalysis_gender_df = pd.merge(purchasinganalysis_gender_df, avgtotalpurchaseperperson, on="Gender", how="outer")

#rename columns
purchasinganalysis_gender_df.columns = ['Purchase Count','Average Purchase Price','Total Purchase Value','Avg Total Purchase per Person']

#print results
purchasinganalysis_gender_df
```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
#age demographics
#remove duplicatate SN to get unique total players
age_demo_df = hop.copy()
age_demo_df = age_demo_df.drop_duplicates(subset='SN', keep="first")

#create bins
bins = [0, 9, 14, 19, 24, 29, 34, 39, 100]
age_group = ['<10', '10-14', '15-19', '20-24', '25-29', '30-34', '35-39', '40+']
```

```python
#age demographics
#bin data by age
age_demo_df['Age Group'] = pd.cut(age_demo_df.Age, bins, labels=age_group)
```

```python
#age demographics
#bin data by age
age_demo_count = age_demo_df["Age Group"].value_counts()
age_demo_percentage = age_demo_count / totalplayers * 100

#display results
new_age_demo_df = pd.DataFrame({"Total Count":age_demo_count, "Percentage of Players":age_demo_percentage})
new_age_demo_df.sort_index()
```

## Purchasing Analysis (Age)

* Bin the purchase_data data frame by age


* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
#purchase analysis (age)
#bin data by age group
hop['Age Group'] = pd.cut(hop.Age, bins, labels=age_group)

#get purchases by age group
hop_age_count = hop["Age Group"].value_counts()

#calculate purchase count, avg purchase price, total purchase value, and avg total purchase per person
purchasecount_age = hop.groupby(['Age Group']).Purchase_ID.count()
avgpurchaseprice_age = hop.groupby(['Age Group']).Price.mean()
totalpurchasevalue_age = hop.groupby(['Age Group']).Price.sum()
avgtotalpurchaseperperson_age = totalpurchasevalue_age / hop_age_count
```

```python
#purchase analysis (age)
#merge groupby into df
purchasinganalysis_age_df = pd.merge(purchasecount_age, avgpurchaseprice_age, on="Age Group", how="outer")
purchasinganalysis_age_df = pd.merge(purchasinganalysis_age_df, totalpurchasevalue_age, on="Age Group", how="outer")

###CANT GET THIS TO MERGE###
#purchasinganalysis_age_df = pd.merge(purchasinganalysis_age_df, avgtotalpurchaseperperson_age, on="Age Group", how="outer", left_index=True)
############################
```

```python
#purchase analysis (age)
#rename columns
purchasinganalysis_age_df.columns = ['Purchase Count','Average Purchase Price','Total Purchase Value']

#print partial results
purchasinganalysis_age_df
```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame

```python
#top spenders (ts)
#find purchase count, avg purchase price, and total purchase value
purchasecount_ts = hop.groupby('SN').Purchase_ID.count()
avgpurchaseprice_ts = hop.groupby('SN').Price.mean()
totalpurchasevalue_ts = hop.groupby('SN').Price.sum()
```

```python
#top spenders
#merge groupby into df
top_spenders_df = pd.merge(purchasecount_ts, avgpurchaseprice_ts, on="SN", how="outer")
top_spenders_df = pd.merge(top_spenders_df, totalpurchasevalue_ts, on="SN", how="outer")

#rename columns
top_spenders_df.columns = ['Purchase Count','Average Purchase Price','Total Purchase Value']

#sort and print final results
top_spenders_df.sort_values(["Total Purchase Value"], ascending=False).head()
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, average item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame

```python
#most popular items
#retrieve item id, item name, and price
hop_items_df = hop[['Item_ID', 'Item_Name', 'Price']].copy()

#groupby item id and item name
item_group = hop_items_df.groupby(['Item_ID', 'Item_Name'])

#find purchase count, item price, and total purchase value
purchasecount_items = item_group['Price'].count()
totalpurchasevalue_items = item_group['Price'].sum()
itemprice = totalpurchasevalue_items / purchasecount_items
```

```python
#most popular items
#merge groupby into df (merge drops item name??? /// unexpected EOF while parsing)
mostpopularitems_df = pd.DataFrame({"Purchase Count" : purchasecount_items, 
                                   "Item Price" : itemprice,
                                   "Total Purchase Value" : totalpurchasevalue_items})

#sort and print final results
mostpopularitems_df.sort_values(["Purchase Count"], ascending=False).head()
```

```python
#most profitable items
mostpopularitems_df.sort_values(["Total Purchase Value"], ascending=False).head()
```
